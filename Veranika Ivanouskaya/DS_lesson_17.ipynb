# Тест на нормальность распределения

```python
import numpy as np
from scipy.stats import shapiro, kstest
```

```python
def normality_tests(dataframe):
    results = {}
    for column in dataframe.select_dtypes(include=['float64', 'int64']).columns:
        data = dataframe[column].dropna()
        
        if len(data) > 5000:
            shapiro_stat, shapiro_p = shapiro(data.sample(5000, random_state=1))
        else:
            shapiro_stat, shapiro_p = shapiro(data)
        
        results[column] = {
            'Shapiro-Wilk Test': {'Statistic': shapiro_stat, 'p-value': shapiro_p},
        }
    
    return results
```

```python
test_results = normality_tests(df)
```

```python
for column, result in test_results.items():
    print(f"Column: {column}")
    print(f"  Shapiro-Wilk Test: Statistic = {result['Shapiro-Wilk Test']['Statistic']}, p-value = {result['Shapiro-Wilk Test']['p-value']}")
```

###### Результаты тестов на нормальность числовых колонок показывают, что p-значения для теста Шапиро-Уилка  близки к нулю. Распределение данных в колонках не соответствует нормальному распределению

## Масштабирование и стандартизация данных
